```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

elections = pd.read_csv('/kaggle/input/elections-csv/elections.csv')
elections.head(5)
```

# **Using loc**

**1. Extract all rows for the candidate 'Andrew Jackson' and display the columns 'Year', 
'Popular vote', and 'Result'.**

Solution: 

```python
elections.loc[elections.loc[:, 'Candidate']=='Andrew Jackson', ['Year', 'Popular vote', 'Result']]
```

**2. Use `loc` to extract all rows where the candidate's `Party` is 'Democratic' and display the 'Year', 'Candidate', and 'Result'.**

Solution: 

```python
elections.loc[elections.loc[:, 'Party']=='Democratic', ['Year', 'Candidate', 'Result']]
```

**3. Extract rows where the `Popular vote` is greater than 600000 and display the 'Candidate' and 'Party' columns.**

Solution: 

```python
elections.loc[elections.loc[:, 'Popular vote'] > 600000, ['Candidate', 'Party']]
```

**4. Use `loc` to find all candidates who won the election and display their 'Year' and 'Candidate'.**

Solution: 

```python
elections.loc[elections.loc[:, 'Result']=='win', ['Year', 'Candidate']]
```

**Extract the 'Year', 'Candidate', and 'Percentage' for candidates from the 'National Republican' party using `loc`.**

Solution: 

```python
elections.loc[elections.loc[:, 'Party']=='National Republican', ['Year', 'Candidate', '%']]
```

# **Combined usage of loc, iloc, and []**

**16. Using `loc`, filter rows where the candidate's `Party` is 'Democratic' and display the first 5 rows of 'Year', 'Candidate', and 'Percentage'.**

**17. Using `iloc`, extract the first 5 rows and display 'Candidate' and 'Party' where the candidate's percentage is above 55.**

Solution: 

* elections.iloc[:, 5]>55 gives the boolen array
* .to_numpy() convert this boolen array into numpy array
* .nonzero method returns 'true' indices from this numpy array
* [0] notation extration the first column from the return result to only get numbers
                all of this required if we want to solve the query using iloc ONLY

```python
elections.iloc[(elections.iloc[:, 5]>55).to_numpy().nonzero()[0], [1,2]]
```
