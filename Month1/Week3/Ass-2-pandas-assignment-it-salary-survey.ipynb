**3. Column Operations and Transformation:**

*a) Create a new column named `Experience Difference`, which shows the difference between `Total years of*
*experience` and `Years of experience in Germany`.*


```python
df['Total years of experience'] = pd.to_numeric(df['Total years of experience'], errors='coerce')
df = df.dropna(subset=['Total years of experience']).copy()
nan_values = df['Total years of experience'].isnull().values.sum()
# print(nan_values)

df['Years of experience in Germany'] = pd.to_numeric(df['Years of experience in Germany'], errors='coerce')
df = df.dropna(subset=['Years of experience in Germany']).copy()
nan_values = df['Years of experience in Germany'].isnull().values.sum()
# print(nan_values)

df['Experience Difference'] = df['Total years of experience'] - df['Years of experience in Germany']
df.loc[:, ['Experience Difference', 'Total years of experience', 'Years of experience in Germany']]
```

*b) Calculate the average salary for each City.*

```python
df['Yearly bonus + stocks in EUR'] = pd.to_numeric(df['Yearly bonus + stocks in EUR'], errors='coerce')
df = df.dropna(subset=['Yearly bonus + stocks in EUR']).copy()
nan_values = df['Yearly bonus + stocks in EUR'].isnull().values.sum()
print(nan_values)

df.groupby('City').agg({'Yearly bonus + stocks in EUR': 'mean'})

```
