```python
# 1. Load the Dataset Load the dataset from the CSV file elections.csv using pandas.read_csv().
elections = pd.read_csv('/kaggle/input/elections-csv/elections.csv')
elections
```

```python
# 2. Basic Exploration How many rows and columns does the dataset have? (Hint: Use shape)
elections.shape
```

```python
# 3. Selecting Data with []

# • Display the list of candidates from the dataset. (Hint: Select the Candidate column) 
elections['Candidate']
```

```python
# 3. Selecting Data with []

# • Display all data for the election year 1828.

# boolenArray = elections['Year'] == 1828
# elections[boolenArray]

elections[elections['Year'] == 1828]
```

```python
# 4. Selecting Data with loc

# • Use loc to display all the rows where the candidate's party is "Democratic."
elections[elections['Party'] == 'Democratic']
```

```python
# 4. Selecting Data with loc
# • Use loc to select the columns 'Year' and 'Popular vote' for elections where the result is 'win.'
elections.loc[elections['Result']=='win', ['Year', 'Popular vote']]
```

```python
# 4. Selecting Data with loc
# • Use loc to find all elections where the percentage of the popular vote was above 55%.
elections.loc[elections['%']> 55, ['Candidate', 'Year', '%']]
```

```python
# 5. Selecting Data with iloc
# • Display the first 5 rows of the dataset using iloc.
elections.iloc[:5]
```

```python
# 5. Selecting Data with iloc
# • Use iloc to select the data from the second to the fourth row (inclusive) and the first three columns.
elections.iloc[2:5, :3]
```

```python
# 5. Selecting Data with iloc
# • Using iloc, extract the 'Popular vote' and 'Result' columns for the first 10 rows.
elections.iloc[:10, [3, 4]]
```

```python
# 6. Combining Conditions with loc and []
# • Display all rows where the result was 'loss' and the percentage of votes was less than 50%.
elections.loc[(elections['Result']=='loss') & (elections['%'] < 50)]
```

```python
# 6. Combining Conditions with loc and []
# • Use both loc and [] to display the candidates from the 'Democratic-Republican' party who won an election.
elections.loc[(elections['Party']=='Democratic-Republican') & (elections['Result']=='win')]
```

```python
# 7. Bonus Challenge
# • Create a new column called 'Vote Margin' that calculates the difference in votes between the candidate and the average popular vote for that election year. Then, display the top 5 candidates with the largest vote margin.

gk = elections.groupby(elections['Year'])
avg_vote = gk['Popular vote'].mean()

elections['Average vote'] = elections['Year'].map(avg_vote)
elections['Vote Margin'] = abs(elections['Average vote'] - elections['Popular vote'])

sort_margin = elections.sort_values('Vote Margin', ascending=False)
sort_margin.head(5)
```
