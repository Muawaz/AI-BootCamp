```python
# 7. Bonus Challenge
# • Create a new column called 'Vote Margin' that calculates the difference in votes between the candidate
# and the average popular vote for that election year. 
# Then, display the top 5 candidates with the largest vote margin.


# elections.loc[elections['Candidate']=='Andrew Jackson']
# print(elections.dtypes)

gk = elections.groupby(elections['Year'])
avg_vote = gk['Popular vote'].mean()
# print(avg_vote)

# elections[['Year', 'Popular vote']]

elections['Average vote'] = elections['Year'].map(avg_vote)
elections['Vote Margin'] = abs(elections['Average vote'] - elections['Popular vote'])
sort_margin = elections.sort_values('Vote Margin', ascending=False)
sort_margin.head(5)

# avg_vote[elections['Year']] - elections['Year']
# elections['Year']

# year_sort = elections.sort_values('Year')
# year_sort
# gk = elections.groupby('Year')
# gk.get_group(1856)
# dic = {}
# elections[['Popular vote', 'Result']]


```
