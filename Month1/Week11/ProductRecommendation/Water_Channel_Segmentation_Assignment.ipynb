```python
from tensorflow.keras.layers import SeparableConv2D, Conv2D
from tensorflow.keras.initializers import HeNormal
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, MaxPooling2D, Dropout, UpSampling2D, concatenate
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
import numpy as np

Beta=[0.011676873, 0.869083715, 0.119239412]
Beta= np.array(Beta)
# print(Beta)
def convert_to_logits(y_pred):
      # see https://github.com/tensorflow/tensorflow/blob/r1.10/tensorflow/python/keras/backend.py#L3525
    y_pred = tf.clip_by_value(y_pred, tf.keras.backend.epsilon(), 1 - tf.keras.backend.epsilon())
    return tf.math.log(y_pred / (1 - y_pred))

def blance_loss(y_true, y_pred):
    y_pred = convert_to_logits(y_pred)
    pos_weight = Beta
    try:
        loss = tf.nn.weighted_cross_entropy_with_logits(logits=y_pred, labels=y_true, pos_weight=pos_weight)
        return tf.reduce_mean(loss * pos_weight)
    except Exception as e:
        print(f"Error in loss calculation: {e}")
        return tf.constant(0.0)

    # or reduce_sum and/or axis=-1
    return tf.reduce_mean(loss *(Beta))

# Model architecture
inputs = Input((256,256,3))
conv1 = SeparableConv2D(64, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(inputs)
conv1 = SeparableConv2D(64, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(conv1)
pool1 = MaxPooling2D(pool_size=(2, 2))(conv1)
conv2 = SeparableConv2D(128, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(pool1)
conv2 = SeparableConv2D(128, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(conv2)
pool2 = MaxPooling2D(pool_size=(2, 2))(conv2)
conv3 = SeparableConv2D(256, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(pool2)
conv3 = SeparableConv2D(256, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(conv3)
pool3 = MaxPooling2D(pool_size=(2, 2))(conv3)
conv4 = SeparableConv2D(512, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(pool3)
conv4 = SeparableConv2D(512, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(conv4)
drop4 = Dropout(0.5)(conv4)
pool4 = MaxPooling2D(pool_size=(2, 2))(drop4)
conv5 = SeparableConv2D(1024, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(pool4)
conv5 = SeparableConv2D(1024, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(conv5)
drop5 = Dropout(0.5)(conv5)
up6 = Conv2D(512, 2, activation='relu', padding='same', kernel_initializer='he_normal')(UpSampling2D(size=(2, 2))(drop5))
merge6 = concatenate([drop4, up6], axis=3)
conv6 = SeparableConv2D(512, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(merge6)
conv6 = SeparableConv2D(512, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(conv6)
up7 = Conv2D(256, 2, activation='relu', padding='same', kernel_initializer='he_normal')(UpSampling2D(size=(2, 2))(conv6))
merge7 = concatenate([conv3, up7], axis=3)
conv7 = SeparableConv2D(256, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(merge7)
conv7 = SeparableConv2D(256, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(conv7)
up8 = Conv2D(128, 2, activation='relu', padding='same', kernel_initializer='he_normal')(UpSampling2D(size=(2, 2))(conv7))
merge8 = concatenate([conv2, up8], axis=3)
conv8 = SeparableConv2D(128, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(merge8)
conv8 = SeparableConv2D(128, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(conv8)
up9 = Conv2D(64, 2, activation='relu', padding='same', kernel_initializer='he_normal')(UpSampling2D(size=(2, 2))(conv8))
merge9 = concatenate([conv1, up9], axis=3)
conv9 = SeparableConv2D(64, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(merge9)
conv9 = SeparableConv2D(64, 3, activation='relu', padding='same', depthwise_initializer='he_normal')(conv9)
conv9 = SeparableConv2D(3, 3, activation='softmax', padding='same', depthwise_initializer='he_normal')(conv9)

model = Model(inputs, conv9)
model.compile(optimizer=Adam(learning_rate=1e-4), loss=blance_loss, metrics=['accuracy'])

# # Summary of the model
# model.summary()
```

```python
# prompt: write an image generator function to train a model

import os
import cv2
import numpy as np
import tensorflow as tf
from google.colab import drive

def image_generator(image_filenames, mask_filenames, image_directory, mask_directory, batch_size, image_size=(256, 256), num_classes=3):
  """
  Generates batches of images and labels from a directory.

  Args:
    image_directory: Path to the directory containing images and corresponding masks.
    batch_size: The size of each batch.

  Yields:
    A tuple containing a batch of images and a batch of corresponding masks.
  """

  # image_filenames = [f for f in os.listdir(image_directory) if f.endswith('.png')] # example, adjust if necessary

  num_images = len(image_filenames)
  print("num_images: ", num_images)

  while True:  # Infinite loop to allow continuous data generation
    for offset in range(0, num_images, batch_size):
      print("num_images: ", num_images)
      batch_image_filenames = image_filenames[offset:offset+batch_size]
      batch_mask_filenames = mask_filenames[offset:offset+batch_size]
      images = []
      masks = [] # assuming you have masks corresponding to images


      for img_filename, mask_filename in zip(batch_image_filenames, batch_mask_filenames):
        try:
          # Load images and corresponding masks
          # Replace 'image_path' and 'mask_path' with your actual filepaths
          image_path = os.path.join(image_directory, img_filename)
          mask_path = os.path.join(image_directory, mask_filename) # Example for mask naming
          # image_path = image_filenames[filename]
          # mask_path = mask_filenames[filename]


          # Check if the file exists
          if not os.path.exists(image_path):
              print(f"Image file does not exist: {image_path}")
              continue
          if not os.path.exists(mask_path):
              print(f"Mask file does not exist: {mask_path}")
              continue

          # Load image using your preferred method (e.g., OpenCV, Pillow)
          img = cv2.imread(image_path)
          if img is None:
            print(f"Failed to load image: {image_path}")
            continue
          img = cv2.resize(img, (256, 256)) # Resize images
          images.append(img)

          # Assuming you have a separate mask for each image
          # Load mask with your preferred method and resize similarly

          # Example: If masks are in a separate directory:
          mask = cv2.imread(mask_path, cv2.IMREAD_GRAYSCALE)
          if mask is None:
            print(f"Failed to load mask: {mask_path}")
            continue  # Skip mask if loading fails
          mask = cv2.resize(mask, image_size)
          # Ensure mask values are within the correct range (0 to num_classes-1)
          mask = np.clip(mask, 0, num_classes - 1)
          masks.append(mask)

          #  # One-hot encode the mask
          # mask_one_hot = np.zeros((image_size[0], image_size[1], num_classes))  # Assuming num_classes is 3
          # for c in range(num_classes):
          #     mask_one_hot[..., c] = (mask == c).astype(int)  # One-hot encoding for each class
          # masks.append(mask_one_hot)
        except Exception as e:
            print(f"Error loading {img_filename}: {e}")
            continue #Skip current image file if an error occurs

      if len(images) == 0 or len(masks) == 0:
        continue  # Skip batch if no valid images/masks are loaded

      # Convert lists to NumPy arrays and normalize if needed.
      # Images:
      images = np.array(images)
      images = images / 255.0 # Normalize pixel values to 0-1

      #Masks:
      masks = np.array(masks)
      # masks = masks / 255.0 #Normalize mask pixel values if necessary
      one_hot_masks = tf.keras.utils.to_categorical(masks, num_classes=num_classes) #convert to one hot encoding if you need

      # yield images, one_hot_masks #yield normalized images and corresponding one-hot encoded masks


      print(f"Yielding batch: {images.shape}, {one_hot_masks.shape}")  # Debug print
      # If no masks are used, simply return images:
      yield images, one_hot_masks


# # Example usage:
# drive.mount('/content/drive')

# # Replace with the actual path to your image directory
# image_dir = '/content/drive/MyDrive/5000 Sementic Segmentation/new_x_' # example path, change to your data
# mask_dir = '/content/drive/MyDrive/5000 Sementic Segmentation/new_y_' # example path, change to your data

# batch_size_ = 2 # Set your desired batch size

# train_generator = image_generator(image_dir, mask_dir, batch_size_)

# # Example of how to use the generator:
# for i in range(10):  # Process 10 batches
#   batch_images, batch_masks = next(train_generator)
#   # Your model training logic here using batch_images and batch_masks
#   print(batch_images.shape, batch_masks.shape)

# # batch_images = next(train_generator)
# # # Your model training logic here using batch_images and batch_masks
# # print(batch_images.shape)
```

```python
import os
import numpy as np
import cv2
from sklearn.metrics.pairwise import cosine_similarity
from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model

base_model = VGG16(weights='imagenet', input_shape=(224, 224, 3), include_top=False)
model = Model(inputs=base_model.input, outputs=base_model.output)
```

```python
drive.mount('/content/drive')

# Replace with the actual path to your image directory
image_dir = '/content/drive/MyDrive/Train_Virual_Try_Images' # example path, change to your data
# image_path = os.path.join(image_dir, 'product_1.jpg')

def extract_feature(image_path):
  img = image.load_img(image_path, target_size=(224, 224))
  img_array = image.img_to_array(img)
  img_array = np.expand_dims(img_array, axis=0)
  img_array = preprocess_input(img_array)
  features = model.predict(img_array)
  features = features.flatten()
  return features

# model.predict()
```

```python
def data_loader(image_directory, image_filenames):
  features_list = []
  for img in image_filenames:
    image_path = os.path.join(image_directory, img)
    features = extract_feature(image_path)
    features_list.append((image_path, features))

  return features_list
```

```python
image_filenames = [f for f in os.listdir(image_dir) if f.endswith('.jpg')]
print(len(image_filenames))
```

```python
features_list = data_loader(image_dir, image_filenames)
```

```python
test_image_dir = '/content/drive/MyDrive/Test_Virual_Try_Images' # example path, change to your data
test_image_path = os.path.join(test_image_dir, 'product_1.jpg')

test_feature = extract_feature(test_image_path)
```

```python

def find_similar_images(test_feature, features_list):
  similarities = []

  for image_path, feature in features_list:
    similarity = cosine_similarity([test_feature], [feature])[0][0]
    similarities.append((image_path, similarity))

  similarities.sort(key=lambda x: x[1], reverse=True)

  return similarities
```

```python
similar_images = find_similar_images(test_feature, features_list)

print(similar_images)
# print(features_list)
```
